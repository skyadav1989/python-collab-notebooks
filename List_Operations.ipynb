### Removing elements from a list by index

There are two primary ways to remove an element from a list using its index:

1.  **`list.pop(index)`**: This method removes the item at the given index and returns it. If no index is specified, `pop()` removes and returns the last item in the list.
2.  **`del list[index]`**: This statement removes the item at the given index. It does not return the removed item.

```python
nums = [2,4,6,8,9,10]
print(nums)
item = nums.pop(1)
print("Item removed"+str(item))
print(f"Item removed {item}")
print("Item removed {it}".format(it=item))
print(nums)
nums.pop()
print("After pop")
print(nums)
del nums[1]
print("After del")
print(nums)
#del nums[10]
```

# Python List Cheatsheet

Lists are ordered, mutable (changeable) sequences of items. They are one of the most versatile data types in Python.

## 1. Creating Lists

```python
# Empty list
my_list = []
print(f"Empty list: {my_list}")

# List of integers
numbers = [1, 2, 3, 4, 5]
print(f"List of numbers: {numbers}")

# List of mixed data types
mixed_list = [1, "hello", 3.14, True]
print(f"Mixed list: {mixed_list}")

# Using list() constructor
list_from_string = list("Python")
print(f"List from string: {list_from_string}")

list_from_tuple = list((10, 20, 30))
print(f"List from tuple: {list_from_tuple}")
```

## 2. Accessing Elements (Indexing & Slicing)

Lists are zero-indexed.

```python
my_list = ['a', 'b', 'c', 'd', 'e']

# Accessing by index
print(f"First element: {my_list[0]}")       # 'a'
print(f"Last element: {my_list[-1]}")       # 'e'

# Slicing: list[start:end:step]
print(f"Elements from index 1 to 3: {my_list[1:4]}") # ['b', 'c', 'd']
print(f"Elements from beginning to index 2: {my_list[:3]}") # ['a', 'b', 'c']
print(f"Elements from index 2 to end: {my_list[2:]}")   # ['c', 'd', 'e']
print(f"All elements (copy): {my_list[:]}")      # ['a', 'b', 'c', 'd', 'e']
print(f"Every other element: {my_list[::2]}")   # ['a', 'c', 'e']
print(f"Reversed list: {my_list[::-1]}")     # ['e', 'd', 'c', 'b', 'a']
```

## 3. Modifying Lists

### Changing an element

```python
my_list = [10, 20, 30, 40, 50]
print(f"Original list: {my_list}")

my_list[1] = 25
print(f"List after changing element at index 1: {my_list}")
```

### Adding elements

```python
my_list = [1, 2, 3]
print(f"Original list: {my_list}")

# append(): Adds an element to the end of the list
my_list.append(4)
print(f"After append(4): {my_list}")

# insert(index, element): Inserts an element at a specified index
my_list.insert(1, 1.5)
print(f"After insert(1, 1.5): {my_list}")

# extend(iterable): Adds elements from an iterable (e.g., another list) to the end
my_list.extend([5, 6])
print(f"After extend([5, 6]): {my_list}")

# Concatenation (+) operator
new_list = my_list + [7, 8]
print(f"After concatenation with [7, 8]: {new_list}")
```

### Removing elements

```python
my_list = [1, 2, 3, 2, 4]
print(f"Original list: {my_list}")

# remove(value): Removes the first occurrence of a specified value
my_list.remove(2)
print(f"After remove(2): {my_list}")

# pop(index): Removes and returns the element at a specified index (defaults to last element if no index is given)
removed_item = my_list.pop(1)
print(f"After pop(1): {my_list}, Removed: {removed_item}")

last_item = my_list.pop()
print(f"After pop() (no index): {my_list}, Removed: {last_item}")

# del statement: Deletes element(s) by index or slice
del my_list[0]
print(f"After del my_list[0]: {my_list}")

my_list = [10, 20, 30, 40, 50]
del my_list[1:3] # Delete elements from index 1 to 2
print(f"After del my_list[1:3]: {my_list}")

# clear(): Removes all elements from the list
my_list.clear()
print(f"After clear(): {my_list}")
```

## 4. List Information

```python
my_list = [1, 2, 3, 2, 4]

# len(): Get the number of elements
print(f"Length of list: {len(my_list)}")

# count(value): Count occurrences of a value
print(f"Count of 2: {my_list.count(2)}")

# index(value): Find the index of the first occurrence of a value
print(f"Index of 3: {my_list.index(3)}")

# Membership testing (in / not in)
print(f"Is 4 in list? {4 in my_list}")
print(f"Is 9 in list? {9 in my_list}")
```

## 5. Iterating Through Lists

```python
my_list = ['apple', 'banana', 'cherry']

# Basic for loop
print("--- Basic for loop ---")
for item in my_list:
    print(item)

# Using enumerate() to get index and value
print("--- Using enumerate() ---")
for index, item in enumerate(my_list):
    print(f"Index {index}: {item}")

# Using range(len()) to get index
print("--- Using range(len()) ---")
for i in range(len(my_list)):
    print(f"Element at index {i}: {my_list[i]}")
```

## 6. Other Useful List Methods

```python
numbers = [3, 1, 4, 1, 5, 9, 2]
print(f"Original numbers: {numbers}")

# sort(): Sorts the list in-place (ascending by default)
numbers.sort()
print(f"After sort(): {numbers}")

# sort(reverse=True): Sorts in descending order
numbers.sort(reverse=True)
print(f"After sort(reverse=True): {numbers}")

# sorted(list): Returns a new sorted list (doesn't modify original)
unsorted_list = [3, 1, 4, 1, 5, 9, 2]
sorted_copy = sorted(unsorted_list)
print(f"Original list: {unsorted_list}, Sorted copy: {sorted_copy}")

# reverse(): Reverses the order of elements in-place
my_list = ['a', 'b', 'c']
my_list.reverse()
print(f"After reverse(): {my_list}")

# copy(): Returns a shallow copy of the list
original = [1, 2, [3, 4]]
copy_list = original.copy()
print(f"Original: {original}, Copy: {copy_list}")

# Modifying nested list in copy affects original (shallow copy)
copy_list[2][0] = 99
print(f"Original after modifying nested in copy: {original}")
```

## 7. List Comprehensions

A concise way to create lists.

```python
# Create a list of squares
squares = [x**2 for x in range(10)]
print(f"Squares: {squares}")

# Filter even numbers
even_numbers = [x for x in range(10) if x % 2 == 0]
print(f"Even numbers: {even_numbers}")

# Nested list comprehension
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flat_list = [num for row in matrix for num in row]
print(f"Flat list: {flat_list}")
```
