Calculates the factorial of a non-negative integer using recursion

```python
def factorial(n):
  """Calculates the factorial of a non-negative integer using recursion."""
  if n == 0:
    return 1
  else:
    return n * factorial(n - 1)

# Example usage:
num = 0
result = factorial(num)
print(f"The factorial of {num} is {result}")
```
