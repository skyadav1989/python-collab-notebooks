## Seaborn Cheatsheet

Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load a sample dataset for demonstration
df = sns.load_dataset('iris')
display(df.head())
```

### 1. Scatter Plot (`scatterplot`)

Used to visualize the relationship between two numerical variables. Can also show the relationship with a third categorical variable using color (`hue`).

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title('Scatter Plot of Sepal Length vs. Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### 2. Line Plot (`lineplot`)

Best for showing trends over time or ordered categories. It connects data points with a line.

```python
# Using a different dataset for line plot example
time_data = sns.load_dataset('fmri')

plt.figure(figsize=(10, 6))
sns.lineplot(x='timepoint', y='signal', hue='region', style='event', data=time_data)
plt.title('fMRI Signal Over Time by Region and Event')
plt.xlabel('Timepoint')
plt.ylabel('Signal Intensity')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

### 3. Bar Plot (`barplot`)

Represents the estimate of central tendency (e.g., mean) for a numerical variable across different categories. Shows confidence intervals by default.

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='species', y='petal_length', data=df, palette='viridis')
plt.title('Average Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length (cm)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### 4. Box Plot (`boxplot`)

Displays the distribution of a numerical variable across different categories, showing quartiles, median, and outliers.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_width', data=df, palette='plasma')
plt.title('Distribution of Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### 5. Histogram (`histplot`)

Shows the distribution of a single numerical variable. You can specify the number of bins or let Seaborn choose.

```python
plt.figure(figsize=(8, 6))
sns.histplot(x='sepal_length', data=df, bins=10, kde=True, hue='species', multiple='stack')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

### 6. Pair Plot (`pairplot`)

Creates a grid of scatterplots for pairs of variables in a dataset, and histograms/KDEs for univariate distributions on the diagonal.

```python
sns.pairplot(df, hue='species')
plt.suptitle('Pair Plot of Iris Dataset by Species', y=1.02) # Adjust title position
plt.show()
```

### 7. Heatmap (`heatmap`)

Visualizes matrix-like data, such as correlation matrices. Each cell represents a value, and its color indicates magnitude.

```python
corr_matrix = df.drop(columns=['species']).corr()

plt.figure(figsize=(8, 7))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Iris Features')
plt.show()
```
