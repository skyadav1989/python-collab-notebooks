Recursion

```python
def fact(n):
  if n<=0 :
    return 1
  return n*fact(n-1)

print(fact(-3))
```

```python
def fact(n):
  if n<=0 :
    return 1
  return n*fact(n-1)

print(fact(977))
#print(fact(978)) #not working max recursion depth reached
```

```python
def fibsum(n):
  if n==0 :
    return 0
  if n==1 :
    return 1
  return fibsum(n-1)+fibsum(n-2)
print(fibsum(5))




```

```python
def fib(n):
  if n<=0 :
    return ""
  first = 0
  second = 1
  if n>=1 :
    print(first,end=" ")
  if n >=2 :
    print(second,end=" ")

  for i in range(2,n):
    next = first+second
    print(next,end=" ")
    first = second
    second = next
  return ""

print(fib(4))
```
